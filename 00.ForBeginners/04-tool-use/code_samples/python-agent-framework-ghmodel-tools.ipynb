```python
# ðŸ› ï¸ Tool Use in AI Agents - Advanced Example
# This notebook demonstrates how to create and use multiple tools with agents

# ðŸ“¦ Import Standard Libraries
import os                     # For environment variable access
from random import randint    # For random number generation

from dotenv import load_dotenv  # For loading environment variables securely
```

```python
# ðŸ¤– Import Agent Framework Components
# These are the core components for building tool-enabled agents
from agent_framework import ChatAgent           # Main agent class
from agent_framework.openai import OpenAIChatClient  # OpenAI-compatible client
```

```python
# ðŸ”§ Load Configuration
# Initialize environment variables for secure API access
load_dotenv()
```

```python
# ðŸŽ² Tool 1: Random Destination Generator
# This is our first tool that provides random travel destinations
# Tools are regular Python functions that agents can call when needed
def get_random_destination() -> str:
    """Get a random vacation destination.
    
    This tool demonstrates basic function calling. The agent can use this
    when users ask for travel suggestions or when planning trips.
    
    Returns:
        str: A randomly selected vacation destination
    """
    # Diverse collection of global destinations
    destinations = [
        "Barcelona, Spain",
        "Paris, France",
        "Berlin, Germany", 
        "Tokyo, Japan",
        "Sydney, Australia",
        "New York, USA",
        "Cairo, Egypt",
        "Cape Town, South Africa",
        "Rio de Janeiro, Brazil",
        "Bali, Indonesia"
    ]
    # Return random selection from the list
    return destinations[randint(0, len(destinations) - 1)]
```

```python
openai_chat_client = OpenAIChatClient(base_url=os.environ.get("GITHUB_ENDPOINT"), api_key=os.environ.get("GITHUB_TOKEN"), ai_model_id=os.environ.get("GITHUB_MODEL_ID"))
```

```python
AGENT_NAME ="TravelAgent"

AGENT_INSTRUCTIONS = """You are a helpful AI Agent that can help plan vacations for customers.

Important: When users specify a destination, always plan for that location. Only suggest random destinations when the user hasn't specified a preference.

When the conversation begins, introduce yourself with this message:
"Hello! I'm your TravelAgent assistant. I can help plan vacations and suggest interesting destinations for you. Here are some things you can ask me:
1. Plan a day trip to a specific location
2. Suggest a random vacation destination
3. Find destinations with specific features (beaches, mountains, historical sites, etc.)
4. Plan an alternative trip if you don't like my first suggestion

What kind of trip would you like me to help you plan today?"

Always prioritize user preferences. If they mention a specific destination like "Bali" or "Paris," focus your planning on that location rather than suggesting alternatives.
"""
```

```python
agent = ChatAgent(
        name = AGENT_NAME,
        chat_client=openai_chat_client,
        instructions=AGENT_INSTRUCTIONS,
        tools=[get_random_destination]
)
```

```python
thread = agent.get_new_thread()
```

```python
response1 = await agent.run("Plan me a day trip",thread= thread)
```

```python

last_message = response1.messages[-1]
text_content = last_message.contents[0].text
print("Travel plan:")
print(text_content)
```

```python
response2 = await agent.run("I don't like that destination. Plan me another vacation.",thread= thread)
```

```python
last_message = response2.messages[-1]
text_content = last_message.contents[0].text
print("Change plan:")
print(text_content)
```
